```python
# A√±adir una columna "Years Playing"
fifa['Years Playing'] = pd.Timestamp.now().year - fifa['Joined']
```

```python
# Filtrar por nacionalidad "Mexico"
mexico_players = fifa[fifa['Nationality'] == 'Mexico']

# Mostrar la informaci√≥n de los jugadores mexicanos
mexico_players
```

```python
# Ordenar por Release Clause
fifa_sorted = fifa.sort_values(by='Release Clause')

# Mostrar los datos ordenados
fifa_sorted
```

```python
# Group by 'Joined' year and count the number of players
players_per_year = fifa.groupby('Joined').size().reset_index(name='Number of Players')

# Display the new DataFrame
players_per_year
```

```python
# Group by 'Joined' year and count the number of players
players_per_year = fifa.groupby('Joined').size().reset_index(name='Number of Players')

# Create the plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.plot(players_per_year['Joined'], players_per_year['Number of Players'], marker='o')
plt.xlabel('Year Joined')
plt.ylabel('Number of Players')
plt.title('Number of Players Joined per Year')
plt.grid(True)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
